> ## What ? 

- Object-oriented programming is an advanced programming paradigm, in which the program is divided into units called objects, methods and classes .


>##  Why ?

- Modularity for easier troubleshooting
- Reuse of code through inheritance
- Flexibility through polymorphism
- simplifying your program.

## Our friend Ahmed works in the Alexandria port as a car dealer. He receives daily shipments of cars and wants to store data for each shipment and performs some operations such as buying and selling cars.

- Data: 
    - model
    - color
    - company
    - number
    - price
- Functions:
    - sell
    - buy
    
> How we can help ahmed ? 


```python
class Car:
    def __init__(self,model,color,company,number,price):
        self.model = model
        self.color = color
        self.company = company
        self.number = number 
        self.price = price
        
    def sell(self,number_of_cars=1):
        self.number -= number_of_cars
    
    def buy(self,number_of_cars=1):
        self.number += number_of_cars
        
    def record(self):
        self.data = {"model":self.model,"Color":self.color,"Company":self.company,"number_of_cars":self.number}
    
    def show_data(self):
        return self.data
        
```

```python
shipment1 = Car("2020","red","BMW",10,50) 
shipment1.record()
shipment1.show_data()
```

```python
shipment1.buy(20)
shipment1.record()
shipment1.show_data()
```

## Ahmed's trade has expanded, so he wants to add a new part to the program, dividing the shipments into sports cars and luxury cars.
- sports cars : with new feature (take back)
- luxury cars : with new feature (warrany "5 Years")

```python
class Car:
    def __init__(self,model,color,company,number,price):
        self.model = model
        self.color = color
        self.company = company
        self.number = number 
        self.price = price
        
    def sell(self,number_of_cars=1):
        self.number -= number_of_cars
    
    def buy(self,number_of_cars=1):
        self.number += number_of_cars
        
    def record(self):
        self.data = {"model":self.model,"Color":self.color,"Company":self.company,"number_of_cars":self.number}
    
    def show_data(self):
        return self.data
        
```

```python
class sport_cars(Car):
    Country = "Europe"
    def take_back(self):
        self.number -= 1 
        
    def warranty(self):
        return "3 Years"
    
class Luxury_cars(Car):
    def warranty(self):
        return "5 Years"
```

## We called that --> Inheritance
- Inheritance is a way of creating a new class for using details of an existing class without modifying it. The newly formed class is a derived class (or child class).

```python
class Car:
    def __init__ (self,model,color,company,number,price):
        self.__model = model
        self.__color = color
        self.__company = company
        self.number = number 
        self.__price = price 
            
    def sell(self,number_of_cars=1):
        self.number -= number_of_cars 
        
    def buy(self,number_of_cars=1):
        self.number += number_of_cars
        
    def record(self):
        self.data = {"Model":self.__model,'color':self.__color,'company':self.__company,'number':self.number}
    
    def show_data(self):
        return self.data 
    
```

```python
class sport(Car):
    Country = "Europe"
    def take_back(self):
        self.number -= 1
    def warranty(self):
        return "3 years"

class luxury(Car):
    country = "German"
    def warranty(self):
        return "5 Years"
```

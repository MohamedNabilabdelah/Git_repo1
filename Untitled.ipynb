```python
import scipy.stats as stats
```

```python
#find the relationship between the categorical data:
#Let's say we want to test if a categorical features is effective for sale price:
#H0: The feature has no effect on sale price.
#H1: The feature has  an effect on sale price.
#If p-value ≤ α: Reject the null hypothesis (H0).
#If p-value > α: Do not reject (H0).
#α:reprisent (F-Statistic).
senif_categ=[]
for categ in feat_categ2:
    groups = [df[df[categ] == val]['SalePrice'] for val in df[categ].unique()]
    anova_result = stats.f_oneway(*groups)
    print(f'---------{categ}-----------')
    print(f"P-value: {anova_result.pvalue} F-Statistic: {anova_result.statistic}")
    print("--"*50)
    if anova_result.pvalue>anova_result.statistic:
        senif_categ.append(categ)
    
#print(f"F-Statistic: {anova_result.statistic}")
#print(f"P-value: {anova_result.pvalue}")
senif_categ
```
